<h3> Importing Dependencies </h3>

```python
import tensorflow as tf
```

<h3> Splitting Data train and test </h3>

```python
dataset=tf.keras.datasets.boston_housing
(x_train,y_train),(x_test,y_test)=dataset.load_data()
```

<h3> Looking at Data </h3>

```python
display(x_train.shape)
display(y_train.shape)
```

<h3> Model creation and Compilation </h3>

```python
model=tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(13,)),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1,activation='linear')
])

model.compile(
    optimizer='adam',
    loss=tf.losses.MeanSquaredError(),
    metrics=['mean_squared_error']
)
```

```python
model.fit(x_train,y_train,epochs=2000,validation_data=(x_test,y_test))
```

```python
display(model.predict(x_test[2].reshape(1,13)))
display(y_test[2])
```
